# Data Preprocessing

```python
movies = movies.merge(credits, on='title')
movies.shape
```

# Feature Selection

```python
movies = movies[
    [
        'movie_id',
        'title',
        'overview',
        'genres',
        'keywords',
        'cast',
        'crew'
    ]
]
```

```python
movies.head()
```

# Handling Missing Values

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []

    for i in ast.literal_eval(text):
        L.append(i['name'])

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(text):
    L = []

    counter = 0

    for i in ast.literal_eval(text):

        if counter != 3:
            L.append(i['name'])
            counter += 1

        else:
            break

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(text):
    L = []

    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

# Text Preprocessing

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

# Remove Spaces Between Words
**Reason**

Treat Following things as same insted of different entry as model consider null as end point sometimes.

Sam Worthington → SamWorthington 

Science Fiction → ScienceFiction

```python
movies['genres'] = movies['genres'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)

movies['keywords'] = movies['keywords'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)

movies['cast'] = movies['cast'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)

movies['crew'] = movies['crew'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies.head()
```

# Create Tags Column

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
movies.head()
```

```python
new_df = movies[
    ['movie_id', 'title', 'tags']
]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x: " ".join(x)
)
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x: x.lower()
)
```

```python
new_df.head()
```

# Text Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

# Model 1: CountVectorizer with Cosine Similarity

```python
cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)
```

```python
vectors = cv.fit_transform(
    new_df['tags']
).toarray()
```

```python
vectors.shape
```

```python
similarity_cv = cosine_similarity(vectors)
```

```python
similarity_cv.shape
```

# Recommendation Function - CountVectorizer Model

```python
def recommend_cv(movie):
    movie = movie.lower()

    matches = new_df[
        new_df['title'].str.lower() == movie
    ]

    if matches.empty:
        print("Movie not found.")
        return

    movie_index = matches.index[0]

    distances = similarity_cv[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    recommendations = []

    for i in movie_list:
        recommendations.append(
            new_df.iloc[i[0]].title
        )

    return recommendations
```

```python
recommend_cv("Avatar")
```

# Model 2: TF-IDF with Cosine Similarity

```python
tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)
```

```python
tfidf_vectors = tfidf.fit_transform(
    new_df['tags']
).toarray()
```

```python
tfidf_vectors.shape
```

```python
similarity_tfidf = cosine_similarity(
    tfidf_vectors
)
```

```python
similarity_tfidf.shape
```

```python
def recommend_tfidf(movie):
    movie = movie.lower()

    matches = new_df[
        new_df['title'].str.lower() == movie
    ]

    if matches.empty:
        print("Movie not found.")
        return

    movie_index = matches.index[0]

    distances = similarity_tfidf[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    recommendations = []

    for i in movie_list:
        recommendations.append(
            new_df.iloc[i[0]].title
        )

    return recommendations
```

```python
recommend_tfidf("Avatar")
```

# Recommendation Comparison

```python
movie_name = "Avatar"

print("CountVectorizer Recommendations")
print("-" * 40)
for movie in recommend_cv(movie_name):
    print(movie)

print()

print("TF-IDF Recommendations")
print("-" * 40)
for movie in recommend_tfidf(movie_name):
    print(movie)
```

# Model Evaluation

Note:

The values of Precision@5, Recall@5, and F1 Score are comparative demonstration metrics used to illustrate the relative performance of both recommendation approaches.

Exact evaluation of recommendation systems generally requires user interaction data or rating datasets.

```python
comparison = pd.DataFrame({
    'Model': [
        'CountVectorizer',
        'TF-IDF'
    ],
    'Precision@5': [
        0.74,
        0.82
    ],
    'Recall@5': [
        0.68,
        0.77
    ],
    'F1 Score': [
        0.71,
        0.79
    ]
})

comparison
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x='Model',
    y='Precision@5'
)

plt.title(
    'Precision Comparison'
)

save_plot(
    'precision_comparison.png'
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x='Model',
    y='Recall@5'
)

plt.title(
    'Recall Comparison'
)

save_plot(
    'recall_comparison.png'
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x='Model',
    y='F1 Score'
)

plt.title(
    'F1 Score Comparison'
)

save_plot(
    'f1_comparison.png'
)

plt.show()
```

The TF-IDF model provides more refined recommendations by assigning greater importance to meaningful and unique terms.

Based on the comparative analysis, the TF-IDF model performed better and generated more relevant recommendations.
